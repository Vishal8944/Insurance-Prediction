```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
ins=pd.read_csv(r"C:\Users\Vishal S\Downloads\archive (9)\insurance.csv")
ins
```

```python
ins.shape
```

```python
ins.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ins['sex']=le.fit_transform(ins['sex'])
ins['smoker']=le.fit_transform(ins['smoker'])
ins['region']=le.fit_transform(ins['region'])
```

```python
ins
```

```python
X=ins.iloc[:,0:6]
Y=ins['expenses']
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=14)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(mse)
print(r2)
```

```python
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print(coefficients)
```

```python
print(f"Intercept: {model.intercept_}")
```

```python
sample_input = np.array([[30, 1, 29.7, 1, 2, 2]])
sample_input = sample_input.reshape(1, -1)

predicted_premium = model.predict(sample_input)
print(f"Predicted Insurance Premium: ${predicted_premium[0]:.2f}")
```
